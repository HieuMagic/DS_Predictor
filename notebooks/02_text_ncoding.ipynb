```python
import pandas as pd

# --- ƒê∆∞·ªùng d·∫´n CSV ---
csv_file = '../data/interim/combined_interim.csv'

def summarize_column(csv_file, column_name, encoding='utf-8-sig'):
    """
    ƒê·ªçc CSV v√† th·ªëng k√™ c·ªôt:
    - S·ªë l∆∞·ª£ng gi√° tr·ªã duy nh·∫•t
    - S·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa t·ª´ng gi√° tr·ªã (bao g·ªìm NaN)
    """
    df = pd.read_csv(csv_file, encoding=encoding)

    if column_name not in df.columns:
        raise ValueError(f"C·ªôt '{column_name}' kh√¥ng t·ªìn t·∫°i trong CSV!")

    counts = df[column_name].value_counts(dropna=False)
    num_unique = df[column_name].nunique(dropna=True)

    print(f"\nüìù C·ªôt '{column_name}' c√≥ {num_unique} gi√° tr·ªã kh√°c nhau (kh√¥ng t√≠nh NaN):\n")
    print(counts)

# --- Danh s√°ch c·ªôt c·∫ßn th·ªëng k√™ ---
fields = ['origin', 'status', 'brand', 'model', 'fuel_type', 'style']

# --- V√≤ng for ---
for col in fields:
    summarize_column(csv_file, col)
```

```python
# Chu·∫©n h√≥a v√† gom nh√£n cho 'status'
status_mapping = {
    'ƒë√£ s·ª≠ d·ª•ng': 'used',
    'xe ƒë√£ d√πng': 'used',
    'xe m·ªõi': 'new',
    'm·ªõi': 'new'
}

combined_df['status_clean'] = (
    combined_df['status']
    .astype(str)          # chuy·ªÉn th√†nh chu·ªói
    .str.lower()          # lowercase ƒë·ªÉ map d·ªÖ d√†ng
    .str.strip()          # lo·∫°i b·ªè kho·∫£ng tr·∫Øng ƒë·∫ßu/cu·ªëi
    .replace(status_mapping)  # gom nh√£n
)
```

```python
# Chu·∫©n h√≥a v√† gom nh√£n cho 'fuel_type'
fuel_mapping = {
    'xƒÉng': 'gasoline',
    'd·∫ßu': 'diesel',
    'ƒëi·ªán': 'electric',
    'hybrid': 'hybrid'
}

combined_df['fuel_clean'] = (
    combined_df['fuel_type']
    .astype(str)
    .str.lower()
    .str.strip()
    .replace(fuel_mapping)
)

print("‚úì ƒê√£ chu·∫©n h√≥a fuel_clean")
print(combined_df['fuel_clean'].value_counts())
```

```python
# One-Hot Encoding cho c·∫£ 2 c·ªôt: status_clean v√† fuel_clean
combined_df = pd.get_dummies(combined_df, columns=['status_clean', 'fuel_clean'], prefix=['status', 'fuel'])

# K·∫øt qu·∫£
print("\n‚úì ƒê√£ √°p d·ª•ng One-Hot Encoding")
print(f"S·ªë c·ªôt sau One-Hot Encoding: {combined_df.shape[1]}")
print(f"\nC√°c c·ªôt m·ªõi:")
print([col for col in combined_df.columns if col.startswith('status_') or col.startswith('fuel_')])
print("\nD·ªØ li·ªáu sau encoding:")
print(combined_df.head())
```
